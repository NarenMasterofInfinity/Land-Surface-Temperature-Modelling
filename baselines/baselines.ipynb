```python
import pandas as pd
from helper import make_madurai_data, load_subset

md = make_madurai_data()

# Discover candidate vars (lightweight)
vars_all = md.vars("madurai")

def choose_day_var(vars_list, prefix):
    cand = [v for v in vars_list if v.startswith(prefix)]
    if not cand:
        return None
    def score(name):
        n = name.lower()
        s = 0
        if "lst" in n or "temp" in n or "temperature" in n:
            s += 4
        if "day" in n:
            s += 3
        if "night" in n:
            s -= 4
        if "qc" in n or "quality" in n or "mask" in n or "valid" in n or "cloud" in n:
            s -= 6
        return s
    return sorted(cand, key=score, reverse=True)[0]

def choose_mask_vars(vars_list, prefix):
    keys = ("qc", "quality", "mask", "cloud", "valid")
    out = []
    for v in vars_list:
        if v.startswith(prefix) and any(k in v.lower() for k in keys):
            out.append(v)
    return out

modis_day = choose_day_var(vars_all, "products/modis/")
viirs_day = choose_day_var(vars_all, "products/viirs/")
modis_masks = choose_mask_vars(vars_all, "products/modis/")
viirs_masks = choose_mask_vars(vars_all, "products/viirs/")

include_vars = [v for v in [modis_day, viirs_day] if v] + modis_masks + viirs_masks
print("MODIS day:", modis_day)
print("VIIRS day:", viirs_day)
print("MODIS masks:", modis_masks)
print("VIIRS masks:", viirs_masks)

# Load only these variables + time slice
ds_c = load_subset(
    md,
    "madurai",
    vars_include=include_vars,
    start="2019-01-01",
    end="2019-12-31",
)
print(ds_c)
```
